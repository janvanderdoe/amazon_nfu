The goal of this code is the merge the links scraped from the category page with the keepa dataset and remove any duplicate variants.

```python
import csv
import pandas as pd
import numpy as np
```

```python
#import all links scraped from the category
links = pd.read_csv("../../gen/scraping/products/amazon_links.csv", sep = ";", header = None)
links.columns = ['title', 'asin', 'link']
```

```python
links
```

```python
#import keepa data to complement link data with parent asin, review count etc.
keepa_data = pd.read_excel("C:/Users/janva/KeepaExport_all_variants.xlsx")
#Select columns
keepa_data = keepa_data[['Title','Reviews: Review Count', 'ASIN', 'Parent ASIN', 'Brand', 'Color']]
#Rename columns
keepa_data.columns = ['title','reviews', 'asin', 'parent_asin', 'brand', 'color']
```

If parent asin is missing, replace it with asin since that asin is the parent. That way, the column contains a list of all the relevant asins (since child asins all fall under the same link

```python
keepa_data.parent_asin.fillna(keepa_data.asin, inplace=True)
```

```python
keepa_data
```

```python
#merge keepa with link dataset
links_keepa = pd.merge(links,keepa_data, on = 'asin', how='left')
```

```python
links_keepa
```

Above dataset contains duplicate links because amazon groups all product variant underneath the parent asin. A quick way to verify this is by checking whether a variant has the same number of reviews as another variant. However, due to a small delay in scraping, there might be a small difference. Hence, first we have to round the number of reviews to the nearest 10.

```python
#Rounding
links_keepa.reviews.fillna(0, inplace=True)
def custom_round(x, base=10):
    return int(base * round(float(x)/base))

links_keepa.reviews = links_keepa.reviews.apply(lambda x: custom_round(x, base=10))

#Now we remove SKUs with the same parent and reviews to avoid scraping those reviews more than once
print(links_keepa[links_keepa.duplicated(subset=['parent_asin', 'reviews'])])
unique_links = links_keepa.drop_duplicates(subset=['parent_asin', 'reviews'], keep='last')
```

```python
#write it to a csv
unique_links.to_csv("../../gen/links/amazon_links_from_category_0_100.csv", sep = ";")
```
